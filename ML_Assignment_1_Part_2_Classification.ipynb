**Step 2:** **Load and View Dataset**

```python
import pandas as pd

df = pd.read_csv("Social_Network_Ads.csv")
df.head()
```

**Step 3:** **Preprocessing**

```python
# Drop User ID

df.drop('User ID', axis=1, inplace=True)
```

```python
# Encode Gender

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
```

```python
# Encode Target Variable

df['Purchased'].value_counts()
```

```python
# Standardize Age & EstimatedSalary

from sklearn.preprocessing import StandardScaler

X = df[['Age', 'EstimatedSalary']]
y = df['Purchased']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

**Step 4: Model Building**

```python
# Test Split

from sklearn.preprocessing import StandardScaler

X = df[['Age', 'EstimatedSalary']]
y = df['Purchased']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#Gaussian Naive Bayes

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.25, random_state=42)

from sklearn.naive_bayes import GaussianNB

nb_model = GaussianNB()
nb_model.fit(X_train, y_train)
y_pred_nb = nb_model.predict(X_test)
```

```python
# K-Nearest Neighbors (try k = 3, 5, 7)

from sklearn.neighbors import KNeighborsClassifier

knn_models = {}
for k in [3, 5, 7]:
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train, y_train)
    knn_models[k] = model.predict(X_test)
```

```python
# Decision Tree (Gini + Entropy)

from sklearn.tree import DecisionTreeClassifier

# Gini
dt_gini = DecisionTreeClassifier(criterion="gini")
dt_gini.fit(X_train, y_train)
y_pred_gini = dt_gini.predict(X_test)

# Entropy
dt_entropy = DecisionTreeClassifier(criterion="entropy")
dt_entropy.fit(X_train, y_train)
y_pred_entropy = dt_entropy.predict(X_test)
```

**Step 6: Evaluation**

```python
# Evaluation of all models using: ■ Accuracy  ■ Precision  ■ Recall  ■ F1-Score  ■ Confusion Matrix

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

def evaluate(y_true, y_pred, model_name):
    print(f"📊 {model_name}")
    print(f"Accuracy: {accuracy_score(y_true, y_pred):.2f}")
    print(f"Precision: {precision_score(y_true, y_pred):.2f}")
    print(f"Recall: {recall_score(y_true, y_pred):.2f}")
    print(f"F1 Score: {f1_score(y_true, y_pred):.2f}")
    print("Confusion Matrix:")
    print(confusion_matrix(y_true, y_pred))
    print("-" * 30)

# Evaluate Naive Bayes
evaluate(y_test, y_pred_nb, "Naive Bayes")

# Evaluate KNNs
for k, preds in knn_models.items():
    evaluate(y_test, preds, f"KNN (k={k})")

# Evaluate Decision Trees
evaluate(y_test, y_pred_gini, "Decision Tree (Gini)")
evaluate(y_test, y_pred_entropy, "Decision Tree (Entropy)")
```

Summary analyzing the performance of **all 3 models**


After running all the models, I compared their performance based on accuracy, precision, recall, and F1-score.

The model that performed the best overall was **KNN with k=7**. It gave the highest F1 score (0.91), along with strong precision and recall. This means it was able to correctly identify the people who purchased and didn’t purchase better than the other models.

Naive Bayes also did well with a high accuracy, but its recall was slightly lower. The Decision Tree models gave lower performance in all the metrics, especially in recall and F1 score.

So, based on the results, I think **KNN (k=7)** was the most accurate model for predicting whether someone would purchase the product or not.
