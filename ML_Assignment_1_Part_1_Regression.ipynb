**Step 1: Data Processing**

```python
# View the Dataset

import pandas as pd

df = pd.read_csv("StudentsPerformance.csv")

df.head()
```

```python
# Check for any missing values in the dataset

df.isnull().sum()
```

```python
# Encode Categorical Variables

df_encoded = pd.get_dummies(df, drop_first=True)

# Show updated dataset
df_encoded.head()
```

**Step 2: Feature Engineering**

```python
from sklearn.preprocessing import StandardScaler

# Target variable
y = df_encoded['math score']

# Feature selection
X = df_encoded.drop(['math score', 'reading score', 'writing score'], axis=1)

# Standardizing the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

**Step 3: Model Building**

```python
# Train-Test Split

from sklearn.model_selection import train_test_split

# Spliting into 80% training and 20% testing
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

**Step 4: Multilinear & Polynomial Regression**

```python
## Train Regression Models
# Multilinear Regression

from sklearn.linear_model import LinearRegression

# Creating and training the model
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

# Predicting
y_pred_linear = lin_reg.predict(X_test)
```

```python
# Polynomial Regression

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Creating polynomial features
poly = PolynomialFeatures(degree=2)  # or 3
X_poly = poly.fit_transform(X_scaled)

# Spliting again with polynomial features
X_train_poly, X_test_poly, y_train_poly, y_test_poly = train_test_split(X_poly, y, test_size=0.2, random_state=42)

# Trainin model
poly_reg = LinearRegression()
poly_reg.fit(X_train_poly, y_train_poly)

# Predicting
y_pred_poly = poly_reg.predict(X_test_poly)
```

**Step 5: Model Evaluation**

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import numpy as np

# Evaluation Function
def evaluate_model(y_true, y_pred, model_name):
    print(f"🔍 {model_name} Performance:")
    print(f"RMSE: {np.sqrt(mean_squared_error(y_true, y_pred)):.2f}")
    print(f"MAE: {mean_absolute_error(y_true, y_pred):.2f}")
    print(f"R² Score: {r2_score(y_true, y_pred):.2f}")
    print("-" * 30)

# Evaluating both models
evaluate_model(y_test, y_pred_linear, "Multilinear Regression")
evaluate_model(y_test_poly, y_pred_poly, "Polynomial Regression")

# Visualization
plt.figure(figsize=(12,5))

# Multilinear
plt.subplot(1,2,1)
plt.scatter(y_test, y_pred_linear, alpha=0.7, color="blue")
plt.xlabel("Actual Math Scores")
plt.ylabel("Predicted")
plt.title("Multilinear Regression")

# Polynomial
plt.subplot(1,2,2)
plt.scatter(y_test_poly, y_pred_poly, alpha=0.7, color="green")
plt.xlabel("Actual Math Scores")
plt.ylabel("Predicted")
plt.title("Polynomial Regression")

plt.tight_layout()
plt.show()
```

Discussion: Overfitting or Underfitting in Polynomial Regression

In my case, the Polynomial Regression model did not perform better than the Multilinear Regression model. It had a lower R² score (0.04) and higher error (RMSE and MAE), which means it was not able to explain much of the variation in math scores.

This shows that the Polynomial Regression model was **underfitting** the data.

Underfitting happens when the model is too simple or not capturing the real patterns in the data
